```python

# 2 numerical & 1 categorical
```

```python
plt.scatter(df['Age'],df['salary'], c=df['dept'].map({"HR":"yellow", "IT":"red","Finance":"blue"}) )
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("AGE vs Salary withh department")
plt.grid()
plt.show()
```

```python
color={"HR":"yellow", "IT":"red","Finance":"blue"}
for dept,color in color.items():
    df_dept=df[df["dept"]==dept]
    plt.scatter(df_dept['Age'],df_dept['salary'], c=color, label=dept)
plt.legend()
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("AGE vs Salary withh department")
plt.grid()
plt.show()
```

# Object Oriented API

```python
fig, axs=plt.subplots(1,3, figsize=(15,5))
## Line plot

axs[0].plot(df['Age'],df['salary'], color='red', marker='o', linestyle=':', linewidth='2')
axs[0].set_title("Age vs Salary")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")
axs[0].grid()

## Histograms
axs[1].hist(df['salary'],bins=10)
axs[1].set_title("Salary Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")
axs[1].grid()

# Boxplot

axs[2].boxplot(df['salary'])
axs[2].set_title("Salary Boxplot")
axs[2].set_xlabel("Salary")
axs[2].grid()
plt.tight_layout()

plt.savefig("multiplePlots.png")
plt.show()
```

3D plot

```python
data = {
    "Salary": [25000, 30000, 37000, 28000, 39000],
    "dept": ["HR", "IT", "Finance", "HR", "Finance"],
    "Age": [22, 25, 29, 24, 30],
    "experience": [1.0, 2.0, 4.0, 1.5, 5.0]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
ax=plt.axes(projection='3d')
ax.scatter(df['Age'],df['Salary'],df['experience'])
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")
plt.show()
```
