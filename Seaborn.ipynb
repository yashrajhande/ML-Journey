```python

# KDE plot
# KDE -> for estimating the probability density function of continuous or non-parametric data
# Or in simple terms: a smooth curve that shows the distribution of data - like a smoothed-out version of histograms
# It helps you see where the data is concentrated (peaks) and how it spreads.

sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", palette = "pastel",fill = "True")
```

```python
# Heatmap:
#  a 2D color-coded matrix that shows relationships or patterns between two variables.
# It's commonly used to:
# - Show correlation between features
# - Visualize confusion matrices -> will see in the later stages.
# - Highlight missing data -> sns.heatmap(df.isnull()>
# woks on only numerical col
```

```python
penguins.head()
```

```python
columns=['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
penguins[columns].corr()
```

```python
sns.heatmap(data=penguins[columns].corr(), annot=True, cmap='Blues')
```

```python
# pair plot
```

```python
sns.pairplot(data=penguins, hue='sex', palette='Set2',diag_kind='kde')
```

```python
# pair grid -> subplot grid for plotting pairwise relationships in a dataset.

graph = sns.PairGrid(data = penguins, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot) #upper triangle
graph.map_lower(sns.kdeplot) #lower triangle
graph.map_diag(sns.histplot) #diagnol
```
